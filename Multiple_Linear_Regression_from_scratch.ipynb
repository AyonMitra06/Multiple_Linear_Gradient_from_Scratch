Now we make the Linear Regression class from scratch

```python
class MultipleRegression:
    def __init__(self):
        self.coef = None
        self.intercept = None

    def fit(self, X_train,y_train):
        # print(X_train.shape) checking if it takes the training data or not
        # now we add a column containg only 1s in the dataset of x train
        X_train = np.column_stack((np.ones(X_train.shape[0]), X_train))
        # print(X_train)

        coeff_matrix = np.linalg.inv((X_train.T).dot(X_train)).dot(X_train.T).dot(y_train)
        # the first index is the intercept and the rest are the coefficients
        self.intercept = coeff_matrix[0]
        self.coef = coeff_matrix[1:]

        print(self.intercept)
        print(self.coef)


    def predict(self, X_test):
        y_pred = np.dot(X_test,self.coef) + self.intercept

        print(r2_score(y_test,y_pred))
        return y_pred
```

```python
reg = MultipleRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.predict(X_test)
```
